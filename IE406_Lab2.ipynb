```
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```
def gradient_descent(init_theta,learning_rate):
  theta_history = np.zeros((10000,init_theta.shape[0]))
  y_list = []
  theta_history[0] = init_theta
  y_list.append(func(init_theta))
  iters = 1
  theta_new = init_theta
  f = False
  it = 10000
  while(True):
    der = derivative(theta_new)
    for i in range(theta_new.shape[0]):
      temp = theta_new[i]
      theta_new[i] = theta_new[i] - learning_rate*der[i] 
      if(abs(temp-theta_new[i])<0.0000001):
        f=True
    theta_history[iters] = theta_new
    y_list.append(func(theta_new))
    iters+=1
    if(iters>=it):
      break
    if(f):
      break
  print("For learning rate ",learning_rate," : ",iters," number of iterations required")
  plt.plot(y_list)
  plt.show()
  return theta_history[:iters]
```

```
def steepest_gradient_descent(init_theta,init_learning_rate):
  theta_history = np.zeros((10000,init_theta.shape[0]))
  y_list = []
  theta_history[0] = init_theta
  y_list.append(func(init_theta))
  iters = 1
  theta_new = init_theta
  learning_rate = init_learning_rate
  f = False
  it = 10000
  alphas = []
  while(True):
    der = derivative(theta_new)
    alphas.append(learning_rate)
    for i in range(theta_new.shape[0]):
      temp = theta_new[i]
      theta_new[i] = theta_new[i] - learning_rate*der[i] 
      if(abs(temp-theta_new[i])<0.0000000001):
          f=True    
    learning_rate = learning_rate - get_derivative(theta_history[iters-1],theta_new,der)
    theta_history[iters] = theta_new
    y_list.append(func(theta_new))
    iters+=1
    if(func(theta_new)==0):
      break
    if(iters>=it):
      break
    if(f):
      break
  print("For learning rate ",init_learning_rate," : ",iters," number of iterations required")
  plt.plot(y_list)
  plt.show()
  return theta_history[:iters]
```

```
def func(x):
  return x**2

def derivative(x):
  return 2*x

x = np.linspace(-2.5,2.5,100)
y = 2

print("Gradient Descent")

theta_init = np.zeros((1))
theta_init[0] = 2

theta_history = gradient_descent(theta_init,0.01)
plt.plot(x,func(x),color = 'orange')
plt.scatter(theta_history,func(theta_history))
plt.show()

theta_init = np.zeros((1))
theta_init[0] = 2

theta_history = gradient_descent(theta_init,0.1)
plt.plot(x,func(x),color = 'orange')
plt.scatter(theta_history,func(theta_history))
plt.show()

def get_derivative(theta_old,theta_new,der,x=x,y=y):
  numerator = -1*theta_new
  denominator = 2*theta_old
  return numerator/denominator

print("Steepest Gradient Descent")

theta_init = np.zeros((1))
theta_init[0] = 2

theta_history = steepest_gradient_descent(theta_init,0.01)
plt.plot(x,func(x),color = 'orange')
plt.scatter(theta_history,func(theta_history))
plt.show()

theta_init = np.zeros((1))
theta_init[0] = 2

theta_history = steepest_gradient_descent(theta_init,0.1)
plt.plot(x,func(x),color = 'orange')
plt.scatter(theta_history,func(theta_history))
plt.show()
```
